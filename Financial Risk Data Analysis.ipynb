```python
risk.columns=['Age', 'Gender', 'Education_Level', 'Marital_Status', 'Income','Credit_Score', 'Loan_Amount', 'Loan_Purpose', 'Employment_Status','Years_at_Current_Job', 'Payment_History', 'Debt_to_Income_Ratio','Assets_Value', 'Number_of_Dependents', 'City', 'State', 'Country','Previous_Defaults', 'Marital_Status_Change', 'Risk_Rating']
```

```python
risk.isnull().sum()
```

```python
risk['Income']=risk['Income'].fillna(np.mean(pd.to_numeric(risk['Income'])))
risk['Credit_Score']=risk['Credit_Score'].fillna(np.mean(pd.to_numeric(risk['Credit_Score'])))
risk['Loan_Amount']=risk['Loan_Amount'].fillna(np.mean(pd.to_numeric(risk['Loan_Amount'])))
risk['Assets_Value']=risk['Assets_Value'].fillna(np.mean(pd.to_numeric(risk['Assets_Value'])))
risk['Number_of_Dependents']=risk['Number_of_Dependents'].fillna(np.mean(pd.to_numeric(risk['Number_of_Dependents'])))
risk['Number_of_Dependents']=risk['Number_of_Dependents'].fillna(np.mean(pd.to_numeric(risk['Number_of_Dependents'])))
```

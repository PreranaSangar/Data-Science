# Seaborn

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
plt.rcParams['figure.figsize'] = 8,4

import warnings
warnings.filterwarnings('ignore')
```

```python
stats.head()
```

```python
vis1 = sns.distplot(stats['InternetUsers'])
```

```python
vis1 = sns.distplot(stats['InternetUsers'], bins=10)
```

```python
vis1 = sns.displot(stats['InternetUsers']) # Histogram
```

```python
vis2 = sns.boxplot(data=stats, x='IncomeGroup', y='BirthRate')
```

```python
vis3 = sns.lmplot(data=stats, x='InternetUsers', y='BirthRate') # Linear model plot
```

```python
vis4 = sns.lmplot(data=stats, x='InternetUsers', y='BirthRate', fit_reg = False) # Removes regression line
```

```python
vis5 = sns.lmplot(data=stats, x='InternetUsers', y='BirthRate', fit_reg=False, hue='IncomeGroup') # hue - parameter for color
```

```python
vis5 = sns.lmplot(data=stats, x='InternetUsers', y='BirthRate', fit_reg=True, hue='IncomeGroup')
```

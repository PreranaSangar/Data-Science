```python
g = sns.JointGrid(x="age", y="bmi", data=medical, height=5, ratio=2)
g = g.plot_joint(sns.kdeplot, cmap="Reds_d")
g = g.plot_marginals(sns.kdeplot, color="r", shade=True)
```

```python
g = sns.JointGrid(x="children", y="charges", data=medical, height=5, ratio=2)
g = g.plot_joint(sns.kdeplot, cmap="magma")
g = g.plot_marginals(sns.kdeplot, color="orange", shade=True)
```

```python
g = sns.JointGrid(x="age", y="charges", data=medical, height=5, ratio=2)
g = g.plot_joint(sns.kdeplot,cmap="mako")
g = g.plot_marginals(sns.kdeplot,fill=True,color="k")
```

```python
sales.columns=['Sales_Person', 'Country', 'Product', 'Date', 'Amount','Boxes_Shipped']
```

```python
sales.head()
```

```python
sales['Amount']=sales['Amount'].str.replace(r'\W','',regex=True) # \W--it removes the non word charachter and that  place it add space
```

```python
sales.head()
```

```python
sales['Sales_Person']=sales['Sales_Person'].astype('category')
sales['Country']=sales['Country'].astype('category')
sales['Product']=sales['Product'].astype('category')
sales['Date']=sales['Date'].astype('category')
sales['Amount']=sales['Amount'].astype('int64')
```
